```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.loadtxt("bestCaseData.txt", usecols=0)
y = np.loadtxt("bestCaseData.txt", usecols=1)

plt.plot(x,y)
plt.title("Best Case")
plt.xlabel("Size of n (number of elements)")
plt.ylabel("Time")
plt.show()
```

```python
x = np.loadtxt("worstCaseData.txt", usecols=0)
y = np.loadtxt("worstCaseData.txt", usecols=1)

plt.plot(x,y)
plt.title("Worst Case")
plt.xlabel("Size of n (number of elements)")
plt.ylabel("Time")
plt.show()
```

```python
x1 = np.loadtxt("worstCaseData.txt", usecols=0)
y1 = np.loadtxt("worstCaseData.txt", usecols=1)
x2 = np.loadtxt("averageCaseData.txt", usecols=0)
y2 = np.loadtxt("averageCaseData.txt", usecols=1)
x3 = np.loadtxt("bestCaseData.txt", usecols=0)
y3 = np.loadtxt("bestCaseData.txt", usecols=1)


plt.plot(x1,y1, color ='r')
plt.plot(x2,y2, color ='g')
plt.plot(x3,y3, color ='y')
plt.title("All Cases")
plt.xlabel("Size of n (number of elements)")
plt.ylabel("Time")
plt.show()
```
